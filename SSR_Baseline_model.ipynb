```python
import numpy as np

y = np.array([10, 12, 14])

ssr_baseline = ((y - y.mean())**2).sum()

print(ssr_baseline)
```

```python
import numpy as np

y = np.array([10, 12, 14])
y_hat_model = np.array([11, 12, 13])

ssr_baseline = ((y - y.mean())**2).sum()
ssr_model = ((y - y_hat_model)**2).sum()

r2 = 1 - (ssr_model / ssr_baseline)

print("SSR baseline:", ssr_baseline)
print("SSR model:", ssr_model)
print("RÂ²:", r2)
```
