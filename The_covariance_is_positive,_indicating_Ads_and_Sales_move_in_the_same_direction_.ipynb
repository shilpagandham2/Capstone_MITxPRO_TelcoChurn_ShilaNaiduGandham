```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("tiny_ads_sales.csv")
df
```

```python
ads = df["Ads"]
sales = df["Sales"]
```

```python
print("Mean Ads:", np.mean(ads))
print("Std Ads:", np.std(ads))

print("Mean Sales:", np.mean(sales))
print("Std Sales:", np.std(sales))
```

```python
plt.scatter(ads, sales)
plt.xlabel("Ads Spend")
plt.ylabel("Sales")
plt.title("Ads vs Sales")
plt.show()
```

```python
print("Covariance Matrix:\n", np.cov(ads, sales))
print("\nCorrelation Matrix:\n", np.corrcoef(ads, sales))
```

```python
corr = np.corrcoef(ads, sales)[0,1]

if corr > 0.8:
    print("\nStrong positive relationship")
elif corr > 0.3:
    print("\nModerate relationship")
else:
    print("\nWeak relationship")
```

```python
np.corrcoef(ads, sales)
```

```python
corr_value = np.corrcoef(ads, sales)[0, 1]

print("Correlation Matrix:\n", np.corrcoef(ads, sales))

print(f"\nCorrelation coefficient: {corr_value:.2f}")

print(
    "\nInterpretation: "
    "Ads spend and Sales exhibit a very strong positive linear relationship."
)
```
